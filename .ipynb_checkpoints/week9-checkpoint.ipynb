```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./prac/Iris.csv")
df.head()
```

```python
x = df.iloc[:,1:3]
x
```

```python
wcss = []
k=list(range(1,11))
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,max_iter=100)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(wcss)
```

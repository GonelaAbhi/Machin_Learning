```python
#1 Import the dataset (Housing.csv) and look at the top five rows
import pandas as pd  
df = pd.read_csv('Housing.csv')
print(df.head())
```

```python
#checking the if any null values in the dataset and there count
print(df.isnull().sum())
```

```python
#fill thr dataset if any null values with 0 
df.fillna(0,inplace=True)#4 Check for missing  values if any, replace with appropriate values
df['price'].fillna(df['price'].mean(), inplace=True)  #replacing by mean value og the column
```

```python
#5 Prepare  X (independent variables) and y (dependent variable)

x = df.drop(columns=['price'])
y = df['price']
```

```python
#6 Visualise the relationship between the independent variables and the depenedent variable using scatterplots

import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x = x["area"],y = y,color="red")
plt.show()
```

```python
x = df.drop(columns=['price'])
y = df['price']
```

```python
import matplotlib.pyplot as plt

num_col = df.select_dtypes(include=['number']).columns
fig,axes = plt.subplots(3,2,figsize=(25,15))
axes=axes.flatten()
#num_col = num_col.drop(columns=['price'])
num_col = num_col.drop('price')
i=0
for col in num_col:
    axes[i].scatter(x=df[col],y=y,label="plot")
    axes[i].set_title(f"{col} graph")
    axes[i].set_xlabel("shp")
    axes[i].legend()
    i+=1

plt.delaxes(axes[-1])
plt.tight_layout()  
plt.show()

```
